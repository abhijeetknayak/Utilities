```python
df.sort_values('Name')
df.sort_values(['Type 1', 'Attack'], ascending=[True, False])
```

```python
# Add Column -----> Total is the new column  FIRST METHOD
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
# df.sort_values("Total", ascending=False)
df = df.drop(columns=['Total'])  # Or 'Total', axis=1 for columns

## SECOND METHOD
df["Total"] = df.iloc[:, 4:10].sum(axis=1)
df.head()

cols = list(df.columns)
print(type(df.columns.values))

print(cols[0:4] + [cols[-1]] + cols[4:12])
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

```python
# Save new df
df.to_csv("new_pokemon.csv", index=False)
df.to_csv("new_pokemon_tabs.txt", index=False, delimiter='\t')
df.to_excel("newpokemon.xlsx")
```
